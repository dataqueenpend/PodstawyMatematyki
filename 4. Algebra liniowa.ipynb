# Algebra liniowa

## Wektory

*Czym jest wektor?*

Wektor to najprostszy obiekt w algebrze liniowej. Wizualizuje się go jako linię ze strzałką zwróconą w określonym kierunku. W najprostszej formie zaczyna się zawsze w punkcie (0,0) układu kartezjańskiego

```python
import matplotlib.pyplot as plt

# Tworzenie wektora
vector = [3, 2]

# Ustalenie początkowych współrzędnych
x_start = 0
y_start = 0

# Obliczenie końcowych współrzędnych
x_end = x_start + vector[0]
y_end = y_start + vector[1]

# Tworzenie wykresu
plt.figure()
plt.quiver(x_start, y_start, vector[0], vector[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Wektor [3, 2], przemierza trzy jednostki w kierunku poziomym i dwie jednostki w kierunku pionowym')
plt.grid()

# Dodawanie widocznych osi x i y
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.show()
```

W pythonie do obliczeń związanych z alegbrą liniową najczęściej używamy NumPy (Numerical Python), który jest wydajniejszy niż zwyczajny Python. Jest to spowodowane tym, że obliczenia w Pythonie są interpretowane dynamicznie, podczas wykonywania. W przeciwieństwie do Pythona, biblioteki takie jak NumPy są napisane w językach nieskopoziomowych (c, c++).

**Po co istnieje wektor?**

W fizyce reprezentuje zwrot i wielkość siły. W matematyce kierunek i skalę na płaszczyźnie XY. **W informatyce jest tablicą służąca do przechowywani danych**.


```python
import matplotlib.pyplot as plt

# Tworzenie wektora
vector = [-3, 2]

# Ustalenie początkowych współrzędnych
x_start = 0
y_start = 0

# Obliczenie końcowych współrzędnych
x_end = x_start + vector[0]
y_end = y_start + vector[1]

# Tworzenie wykresu
plt.figure()
plt.quiver(x_start, y_start, vector[0], vector[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-4, 1)
plt.ylim(-1, 3)
plt.title('Wektor [-3, 2], przemierza trzy jednostki w kierunku poziomym i dwie jednostki w kierunku pionowym')
plt.grid()

# Dodawanie widocznych osi x i y
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.show()
```

Wektory w informatyce używa się do uczenia maszynowego i statystyki, aby praca z liczbami była łatwiejsza. Mają one jednak znacznie szersze zastosowanie m.in. w grafice komputerowej, grach wideo - gdzie przetwarzają grafiki, oddziaływania fizyczne. 

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Tworzenie wektora
vector = np.array([4, 1, 2])

# Ustalenie początkowych współrzędnych
x_start = 0
y_start = 0
z_start = 0

# Obliczenie końcowych współrzędnych
x_end = x_start + vector[0]
y_end = y_start + vector[1]
z_end = z_start + vector[2]

# Tworzenie wykresu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dodawanie linii pomocniczych
ax.plot([x_start, x_end], [y_start, y_start], [z_start, z_start], 'r--')  # Linia pomocnicza dla osi X
ax.plot([x_end, x_end], [y_start, y_end], [z_start, z_start], 'r--')  # Linia pomocnicza dla osi Y
ax.plot([x_end, x_end], [y_end, y_end], [z_start, z_end], 'r--')  # Linia pomocnicza dla osi Z

# Dodawanie wektora
ax.quiver(x_start, y_start, z_start, vector[0], vector[1], vector[2], length=1)

# Ustawianie takich samych wartości liczbowych na osiach
ax.set_xticks(np.arange(0, 6, 1))
ax.set_yticks(np.arange(0, 3, 1))
ax.set_zticks(np.arange(0, 4, 1))

ax.set_xlabel('Oś X')
ax.set_ylabel('Oś Y')
ax.set_zlabel('Oś Z')
ax.set_title('Wektor [4, 1, 2]')
plt.show()
```

Na powyższym rysunku wektora trójwymiarowego: 
* przemierzamy 4 kroki na osi x
* przemierzamy 1 krok na osi y
* przemierzamy 2 kroki na osi z

```python
import matplotlib.pyplot as plt
import numpy as np

# Tworzenie wektorów
vector = np.array([3, 2])
vector_2 = np.array([2, -1])

# Dodawanie wektorów
resultant_vector = vector + vector_2

# Ustalenie początkowych współrzędnych
x_start = 0
y_start = 0

# Obliczenie końcowych współrzędnych
x_end = x_start + vector[0]
y_end = y_start + vector[1]

# Obliczenie końcowych współrzędnych wektora 2
x_end_2 = x_end + vector_2[0]
y_end_2 = y_end + vector_2[1]

# Obliczenie końcowych współrzędnych wektora wynikowego
x_end_resultant = x_start + resultant_vector[0]
y_end_resultant = y_start + resultant_vector[1]

# Tworzenie wykresu
plt.figure()

# Rysowanie wektorów
plt.quiver(x_start, y_start, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector v')
plt.quiver(x_end, y_end, vector_2[0], vector_2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector w')
plt.quiver(x_start, y_start, resultant_vector[0], resultant_vector[1], angles='xy', scale_units='xy', scale=1, color='g', label='v+w')

# Dodawanie widocznych osi x i y
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlim(-1, 4)
plt.ylim(-1, 5)

# Dodawanie etykiet wektorów
plt.text(x_end + 0.2, y_end + 0.2, f'Wektor v: {vector[0]}, {vector[1]}', ha='left', va='top', fontsize=8)
plt.text(x_end_2 + 0.2, y_end_2 + 0.2, f'Wektor w: {vector_2[0]}, {vector_2[1]}', ha='left', va='top', fontsize=8)
plt.text(x_end_resultant + 0.2, y_end_resultant + 0.2, f'Wektor v+w: {resultant_vector[0]}, {resultant_vector[1]}',ha='left', va='top', fontsize=8)

plt.grid()
plt.xlim(-2, 6)
```
